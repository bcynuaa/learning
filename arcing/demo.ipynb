```python
import numpy as np
dt=0.000001
Final=0.2
Range=int(Final/dt)
para={'U':17,'t1':0.5e-3,'theta':3.5e-4,'u_arc':12,'U_alt':115,'f':400,'p0':10,'R':20}
t=np.arange(0,Final-dt,dt)
u_s=para['U_alt']*np.cos(2*np.pi*para['f']*t)
u=np.zeros(Range)
i=u
g=u
g[0]=100
#i[0]=g[0]*u_s[0]

for k in range(Range-1):
    u[k]=u_s[k]-para['R']*i[k]
    g[k]=i[k]/u[k]
    temp=1/para['theta']*(u[k]*i[k]/max(para['u_arc']*np.abs(i[k]),para['p0'])-1)
    g[k+1]=np.exp(temp*dt+np.log(g[k]))
    i[k+1]=g[k+1]*u[k]
    pass
```

```python
import matplotlib.pyplot as plt
plt.plot(t,i)
plt.show()
```
