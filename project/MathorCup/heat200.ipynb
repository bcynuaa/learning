```python
import numpy as np
from matplotlib import pyplot as plt
import math
import heat200 as sys

T_init=sys.get_T_init()
plt.imshow(T_init, cmap='bwr')
```

```python
T_next=sys.next_T(T_init)
plt.imshow(T_next,cmap='bwr')
T_next.max()-T_init.max()
```

```python
T_iter=T_init
Tmax=list()
for i in range(int(6*6*300)):
    Tmax.append(T_iter.max())
    T_iter=sys.next_T(T_iter)
    if Tmax[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax=np.array(Tmax)-sys.T0
print(T_iter.max()-sys.T0)
plt.imshow(T_iter,cmap='bwr')
plt.show()
plt.plot(Tmax)
```

```python
T_iter1=T_iter
Tmax1=list()
for i in range(int(6*6*300)):
    Tmax1.append(T_iter1.max())
    T_iter1=sys.next_T(T_iter1)
    if Tmax1[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax1=np.array(Tmax1)-sys.T0
print(T_iter1.max()-sys.T0)
plt.imshow(T_iter1,cmap='bwr')
plt.show()
plt.plot(Tmax1)
```

```python
T_iter2=T_iter1
Tmax2=list()
for i in range(int(6*6*300)):
    Tmax2.append(T_iter2.max())
    T_iter2=sys.next_T(T_iter2)
    if Tmax2[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax2=np.array(Tmax2)-sys.T0
print(T_iter2.max()-sys.T0)
plt.imshow(T_iter2,cmap='bwr')
plt.show()
plt.plot(Tmax2)
```

```python
TTT=np.r_[Tmax,Tmax1,Tmax2]
ttt=int(TTT.size*100/60/60/24*100)/100
ddd=np.linspace(0,ttt,TTT.size)
TTTconst=np.zeros(TTT.size)+80
plt.plot(ddd,TTT,label='$T_{max}$-day')
plt.plot(ddd,TTTconst,label='$T=80\degree$')
plt.scatter(ddd[-1],80,s=200,color='r',alpha=0.5,marker='^')
plt.text(ddd[-1]-6,73,'$T_{max}>80\degree$',fontsize='x-large',color='r')
plt.legend(fontsize='x-large')
plt.grid(True)
plt.title('$T_{max}$ with Days When Number=200')
plt.xlabel('t / Days')
plt.ylabel('$T_{max}$ / $\degree C$')
plt.savefig('200.png',format='png')
```
