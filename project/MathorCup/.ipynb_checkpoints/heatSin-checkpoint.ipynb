```python
import numpy as np
from matplotlib import pyplot as plt
import math
import heatSin as sys
import seaborn as sns

T_init=sys.get_T_init()
plt.imshow(T_init, cmap='bwr')
```

```python
T_next=sys.next_T_t(T_init,0)
plt.imshow(T_next,cmap='bwr')
T_next.max()-T_init.max()
```

```python
T_iter=T_init
Tmax=list()
for i in range(int(6*6*300*7)):
    Tmax.append(T_iter.max())
    T_iter=sys.next_T_t(T_iter,i)
    if Tmax[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax=np.array(Tmax)-sys.T0
print(T_iter.max()-sys.T0)
plt.imshow(T_iter,cmap='bwr')
plt.show()
plt.plot(Tmax)
```

```python
T_iter1=T_iter
Tmax1=list()
for j in range(int(6*6*300)):
    Tmax1.append(T_iter1.max())
    T_iter1=sys.next_T_t(T_iter1,j)
    if Tmax1[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax1=np.array(Tmax1)-sys.T0
print(T_iter1.max()-sys.T0)
plt.imshow(T_iter1,cmap='bwr')
plt.show()
plt.plot(Tmax1)
```

```python
T_iter2=T_iter1
Tmax2=list()
for k in range(int(6*6*300)):
    Tmax2.append(T_iter2.max())
    T_iter2=sys.next_T_t(T_iter2,k)
    if Tmax2[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax2=np.array(Tmax2)-sys.T0
print(T_iter2.max()-sys.T0)
plt.imshow(T_iter2,cmap='bwr')
plt.show()
plt.plot(Tmax2)
```

```python
T_iter3=T_iter2
Tmax3=list()
for kk in range(int(6*6*300)):
    Tmax3.append(T_iter3.max())
    T_iter3=sys.next_T_t(T_iter3,kk)
    if Tmax3[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax3=np.array(Tmax3)-sys.T0
print(T_iter3.max()-sys.T0)
plt.imshow(T_iter3,cmap='bwr')
plt.show()
plt.plot(Tmax3)
```

```python
TTT=np.r_[Tmax,Tmax1,Tmax2,Tmax3]
ttt=10*6*6*300*100/60/60/24
TTTconst=Tmax1.max()+np.zeros(TTT.size)
ddd=np.linspace(0,ttt,TTT.size)
plt.plot(ddd,TTT,label='$T_{max}$-day')
plt.plot(ddd,TTTconst,label='$T_{final}=$'+str(int(Tmax1.max()*100)/100)+'$\degree C$')
plt.legend(fontsize='x-large',loc='best')
plt.grid(True)
plt.title('$T_{max}$ with Days When Number=160\n Consedering Seasonal Changes',fontsize='x-large')
plt.xlabel('t / Days',fontsize='x-large')
plt.ylabel('$T_{max}$ / $\degree C$',fontsize='x-large')
plt.scatter(95,Tmax1.max(),color='r',s=100,alpha=0.5)
plt.text(85,70,'$T_{top}=79.27\degree C$',color='r',fontsize='x-large')
plt.savefig('Sin.png',format='png')
```
