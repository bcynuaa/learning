```python
import numpy as np
from matplotlib import pyplot as plt
import math
import heat170 as sys
import seaborn as sns

T_init=sys.get_T_init()
plt.imshow(T_init, cmap='bwr')
```

```python
T_next=sys.next_T(T_init)
plt.imshow(T_next,cmap='bwr')
T_next.max()-T_init.max()
```

```python
T_iter=T_init
Tmax=list()
for i in range(int(6*6*300*7)):
    Tmax.append(T_iter.max())
    T_iter=sys.next_T(T_iter)
    if Tmax[-1]>80+sys.T0:
        print('在 '+str(i)+' 步超过80度！')
        break
    pass
Tmax=np.array(Tmax)-sys.T0
print(T_iter.max()-sys.T0)
plt.imshow(T_iter,cmap='bwr')
plt.show()
plt.plot(Tmax)
```

```python
TTT=Tmax
ttt=int(TTT.size*100/60/60/24*100)/100
ddd=np.linspace(0,ttt,TTT.size)
TTTconst=np.zeros(TTT.size)+Tmax[-1]
plt.plot(ddd,TTT,label='$T_{max}$-day')
plt.plot(ddd,TTTconst,label='$T_{final}=79.14\degree$')
plt.scatter(ddd[-1],80,s=200,color='r',alpha=0.5,marker='^')
plt.text(ddd[-1]-22,72,'$T_{final}=79.14\degree$\n Day:8.5',fontsize='x-large',color='r')
plt.legend(fontsize='x-large')
plt.grid(True)
plt.title('$T_{max}$ with Days When Number=170')
plt.xlabel('t / Days')
plt.ylabel('$T_{max}$ / $\degree C$')
plt.savefig('170.png',format='png')
```
