```python
import numpy as np
import matplotlib.pyplot as plt
dP=2000
E=7.06e6
G=2.7e6
d=np.array([40,115,200,300,500,700])
P=np.zeros(8)
for i in range(0,8):
    P[i]=i*dP
    pass
data=np.loadtxt('data.txt')
data=data.T

def get(e0,e45,e_45):
    ex=e0
    ey=e45+e_45-e0
    exy=e45-e_45
    sigmax=E*ex
    tau=G*(e45-e_45)
    return ex,ey,exy,sigmax,tau
    pass

class ceng:
    def __init__(self,e0,e45,e_45,ezhu,ece):
        self.ex,self.ey,self.exy,self.sigmax,self.tau=get(e0,e45,e_45)
        self.e0=e0
        self.e45=e45
        self.e_45=e_45
        self.ezhu=ezhu
        self.sigmazhu=E*ezhu
        self.ece=ece
        self.sigmace=E*ece
        pass
```

```python
data10=data[6]
dat10=list()
for i in range(6):
    tmp_e45=data10[i*3]
    tmp_e0=data10[i*3+1]
    tmp_e_45=data10[i*3+2]
    tmp_ezhu=data10[i+18]
    tmp_ece=data10[i+24]
    tmp=ceng(tmp_e0,tmp_e45,tmp_e_45,tmp_ezhu,tmp_ece)
    dat10.append(tmp)
    pass
```

```python
num=np.array(range(0,30))+1
tongdao=(data[-1]-data[0])*10/14
for i in range(30):
    plt.scatter(num[i],tongdao[i],s=100,alpha=1,color='b')
    pass
plt.grid(which='major')
plt.xlabel('Tunnel ID')
plt.ylabel('$\epsilon\quad\mu\epsilon$')
plt.title('$\Delta P=10kN\quad\mu\epsilon$ - Tunnel ID \n Linear Fitted')
plt.show()
```

# B

```python
Sigmazhu=np.zeros(6)
Sigmace=np.zeros(6)
for i in range(6):
    Sigmazhu[i]=dat10[i].sigmazhu
    Sigmace[i]=dat10[i].sigmace
    pass

#图1
plt.plot(d,Sigmazhu,label='$\sigma_{main}-x$')
plt.grid(True)
plt.legend()
plt.xlabel('x(cm)')
plt.ylabel('$\sigma_{main}(N/cm^2)$')
plt.title('$\Delta P=10kN\quad\sigma_{main}-x$')
plt.show()

#图2
plt.plot(d,Sigmace,label='$\sigma_{edge}-x$')
plt.grid(True)
plt.legend()
plt.xlabel('x(cm)')
plt.ylabel('$\sigma_{edge}(N/cm^2)$')
plt.title('$\Delta P=10kN\quad\sigma_{edge}-x$')
plt.show()
```

```python
Tau=np.zeros(6)
for i in range(6):
    Tau[i]=dat10[i].tau
    pass
plt.plot(d,Tau,label=r'$\tau-x$')
plt.legend()
plt.xlabel('x(cm)')
plt.ylabel(r'$\tau (N/cm^2)$')
plt.title(r'$\Delta P=10kN\quad\tau-x$')
plt.grid(True)
plt.show()
```
