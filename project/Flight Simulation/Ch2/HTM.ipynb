```python
import modsim as ms
m=ms.UNITS.meter
s=ms.UNITS.second
g=9.8*m/s**2
init_z=ms.State(y=100*m,v=0*m/s)
init_x=ms.State(y=0*m,v=200*m/s)
t_end=6*s
dt=0.01*s
system_z=ms.System(init=init_z,g=g,t_end=t_end,dt=dt)
system_x=ms.System(init=init_x,t_end=t_end,dt=dt)
```

```python
def slope_func_z(state,t,system):
    y,v=state
    dydt=v
    dvdt=-system.g
    return dydt,dvdt
    pass
def slope_func_x(state,t,system):
    y,v=state
    dydt=v
    dvdt=0*m/s**2
    return dydt,dvdt
    pass
```

```python
Vz,Az=slope_func_z(system_z.init,0,system_z)
Vx,Ax=slope_func_x(system_x.init,0,system_x)
results_z,details_z=ms.run_euler(system_z,slope_func_z)
results_x,details_x=ms.run_euler(system_x,slope_func_x)
```

```python
from matplotlib import pyplot as plt
ms.plot(results_z.y)
ms.decorate(xlabel='Times',ylabel='Height')
```
