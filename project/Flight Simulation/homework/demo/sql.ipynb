```python
import sqlite3 as sql
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
```

```python
con = sql.connect('sql.db')
cur = con.cursor()
cur.execute("create table demo(x double UNIQUE primary key,y double);")
```

```python
for i in range(x.size):
    data = str(x[i]) + ',' + str(y[i])
    cur.execute('INSERT INTO demo VALUES (%s)'%data)
    pass
con.commit()
con.close()
del con
del cur
del data
```

```python
con = sql.connect('sql.db')
cur = con.cursor()
Data = cur.execute("Select x , y from demo")
```
